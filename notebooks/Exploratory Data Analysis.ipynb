```python
# Load the data
port_df = pd.read_csv("../data/student-por.csv")
port_df.info()
```

```python
port_df.isnull().sum()
```

Convert grade columns to percentages

```python
port_df[['G1_perc', 'G2_perc', 'G3_perc']] = port_df[['G1', 'G2', 'G3']] / 20
```

Create average grade column

```python
port_df['G1G2_mean'] = np.mean(port_df[['G1_perc', 'G2_perc']], axis=1)
```

Start by clustering the numeric data to see if there are patterns of behaviour that are detrimental to a student's performance (study time, grade, failures)

```python
clust_cols = [
  "studytime", 
  "failures", 
  "famrel", 
  "goout", 
  "Dalc", 
  "Walc",
  "health",
  "absences",
  "G1G2_mean",
  "G3_perc"]
```

```python
port_df[clust_cols].describe()
```

* Average study time is about 2 hours
* Failures are uncommon
* Family relationships are good, on average
* Most students have a moderate nightlife
* Not a lot of students drink a lot, when they do they tend to drink more on weekends
* Students tend to be in good health
* Class average is low
* Class absenteism is not really a problem

It seems weird that there are students with 10% average grade over periods 1 and 2 and 0% final grade.  Are these the students that failed?

```python
port_df.query("G3_perc == 0")[["G1", "G2", "G3", "G1G2_mean", "failures"]]
```

At this point, we can assume that student's needed an average of > 50% to pass and receiving a final grade of 0 indicates that they failed the course.

```python
from sklearn import preprocessing
scaler = preprocessing.MaxAbsScaler()

stud_clusters = scaler.fit_transform(port_df[clust_cols])
stud_clusters = pd.DataFrame(stud_clusters, columns=clust_cols)
```

```python
from sklearn.cluster import KMeans

def kmeans_cluster(df, n_clusters=2):
    model = KMeans(n_clusters=n_clusters, random_state=1)
    clusters = model.fit_predict(df)
    cluster_results = df.copy()
    cluster_results['cluster'] = clusters
    return cluster_results
  
def summarize_clustering(results):
    cluster_size = results.groupby(['cluster']).size().reset_index()
    cluster_size.columns = ['cluster', 'count']
    cluster_means = results.groupby(['cluster'], as_index=False).mean()
    cluster_summary = pd.merge(cluster_size, cluster_means, on='cluster')
    return cluster_summary
```

```python
cluster_results = kmeans_cluster(stud_clusters, 8)
cluster_summary = summarize_clustering(cluster_results)
cluster_summary['count']
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.heatmap(cluster_summary[clust_cols].transpose(), annot=True);
```

Similar clusters include:
* 0,2
* 3, 6

Reduce cluster size to get more pure clusters

```python
cluster_results = kmeans_cluster(stud_clusters, 5)
cluster_summary = summarize_clustering(cluster_results)
sns.heatmap(cluster_summary[clust_cols].transpose(), annot=True);
```

```python
cluster_summary['count'] / cluster_summary['count'].sum()
```

Sutdents in the first cluster perform better on average as indicated by the higher mean grades.  Compared to students in other clusters, these students tend to study more, have fewer failures, better family relationships, go out less, drink less during the weekend, and have fewer absences.

Students in the second cluster performs significantly worse, on average, with mean grades below 50%.  These students tend to fail more classes, have poorer quality familial relationships, drink more during the week and weekend, and have more absences.

Students in the third cluster are close to failing, on average, and tend to go out and drink significantly more than students in other clusters.
